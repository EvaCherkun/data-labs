```python
import numpy as np

# Створення скалярного значення
scalar = np.random.randint(10)
print(f"Початкове скалярне значення: {scalar}")

# Створення вектору (одновимірний масив) за допомогою arange
vector = np.arange(1, 6)  # Вектор значень від 1 до 5
print(f"Початковий вектор: {vector}")

# Створення матриці (двовимірний масив) випадкових значень
matrix = np.random.randint(1, 10, size=(3, 3))  # 3x3 матриця з випадкових чисел від 1 до 9
print(f"Початкова матриця:\n{matrix}\n")

# Операції CRUD

#CREATE

print("Створення нових елементів...")

scalar = scalar + 10  # Оновлюємо скаляр
vector = np.append(vector, 99)  # Додаємо елемент до вектора
new_row = np.random.randint(1, 10, size=(1, 3))  # Нова строка для матриці
matrix = np.vstack([matrix, new_row])  # Додаємо новий рядок до матриці


print(f"Оновлений скаляр (після створення): {scalar}")
print(f"Оновлений вектор (після додавання): {vector}")
print(f"Оновлена матриця (після додавання нового рядка):\n{matrix}\n")

#READ

print(f"Читання елемента вектора [2]: {vector[2]}")
print(f"Читання елемента матриці [1, 1]: {matrix[1, 1]}\n")

#UPDATE

print("Оновлення елементів...")
vector[0] = 100  # Змінюємо перший елемент вектора
matrix[0, 0] = 42  # Змінюємо елемент [0, 0] в матриці

print(f"Оновлений вектор (після зміни): {vector}")
print(f"Оновлена матриця (після зміни):\n{matrix}\n")

#DELETE

print("Видалення елементів...")

vector = np.delete(vector, 1)  # Видаляємо другий елемент з вектора
matrix = np.delete(matrix, 2, axis=0)  # Видаляємо третій рядок з матриці

print(f"Оновлений вектор (після видалення): {vector}")
print(f"Оновлена матриця (після видалення рядка):\n{matrix}")
```

```python
import pandas as pd

# Створення DataFrame з NumPy масиву
array_data = np.random.randint(0, 100, size=(5, 3))
df_from_array = pd.DataFrame(array_data, columns=['A', 'B', 'C'])
print("DataFrame з NumPy масиву:")
print(df_from_array)

# Створення DataFrame зі словника
data_dict = {
    'Name': ['Taylor', 'Melovin', 'Peter', 'Linda', 'James'],
    'Age': [34, 27, 35, 32, 29],
    'City': ['New York', 'Kyiv', 'London', 'Berlin', 'Madrid']
}
df_from_dict = pd.DataFrame(data_dict)
print("\nDataFrame зі словника:")
print(df_from_dict)


print("\nПерші 3 рядки DataFrame з NumPy масиву:")
print(df_from_array.head(3))
print("\nПерші 2 рядки DataFrame зі словника:")
print(df_from_dict.head(2))


print("\nОпис статистики для DataFrame з NumPy масиву:")
print(df_from_array.describe())
print("\nОпис статистики для DataFrame зі словника:")
print(df_from_dict.describe())

# Індексація за допомогою iloc
print("\nВибір 2-го рядка та першого стовпця за допомогою iloc:")
print(df_from_array.iloc[1, 0])

# Індексація за допомогою loc
print("\nВибір рядка з іменем 'Linda' та стовпця 'Age' за допомогою loc:")
print(df_from_dict.loc[df_from_dict['Name'] == 'Linda', 'Age'])

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  

# Створення даних
x = np.linspace(0, 20, 15)  # Більший діапазон і більше точок
y = 3 * x - 5  # лінійна функція
y_noise = y + np.random.normal(0, 2, len(x))  

# Побудова графіку
plt.plot(x, y, label='Лінійна функція', linestyle='--', color='blue')
plt.scatter(x, y_noise, color='green', label='Дані з помилками')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Лінійна функція з доданим шумом')
plt.legend()
plt.grid(True)
plt.show()

# Обчислення метрик похибок
mae = np.mean(np.abs(y - y_noise))
mse = np.mean((y - y_noise)**2)
print(f'Середня абсолютна помилка (MAE): {mae}')
print(f'Середньоквадратична помилка (MSE): {mse}')

# Запис результатів у CSV файл
results = pd.DataFrame({'X': x, 'Y': y, 'Y_noise': y_noise})
results['MAE'] = mae
results['MSE'] = mse
results.to_csv('modified_results.csv', index=False)
```
